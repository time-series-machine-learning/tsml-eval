<h1>Paper path and alginment generation</h1>

```python
import warnings

import numpy as np

warnings.filterwarnings(
    "ignore"
)  # Hide warnings that can generate and clutter notebook

x = np.array(
    [
        -0.7553383207,
        0.4460987596,
        1.197682907,
        0.1714334808,
        0.5639929213,
        0.6891222874,
        1.793828873,
        0.06570866314,
        0.2877381702,
        1.633620422,
    ]
)

y = np.array(
    [
        0.01765193577,
        1.536784164,
        -0.1413292622,
        -0.7609346135,
        -0.1767363331,
        -2.192007072,
        -0.1933165696,
        -0.4648166839,
        -0.9444888843,
        -0.239523623,
    ]
)
```

```python
# Find the distance between time series
from aeon.distances import distance

dtw = distance(x, y, metric="dtw")
dtw_windowed = distance(x, y, metric="dtw", window=0.2)
edr = distance(x, y, metric="edr")
erp = distance(x, y, metric="erp", epsilon=1.0)
lcss = distance(x, y, metric="lcss")
msm = distance(x, y, metric="msm")
twe = distance(x, y, metric="twe")
wdtw = distance(x, y, metric="wdtw")
wdtw_g2 = distance(x, y, metric="wdtw", g=0.2)
wdtw_g3 = distance(x, y, metric="wdtw", g=0.3)
```

```python
# If you want to repeatedly use a distance factory so that function is cached (i.e. not recompiled to c every time called)
from aeon.distances import distance_factory

# To use the factory the time series MUST be preconfigured to (d, m) format
x_numba = np.array([x])
y_numba = np.array([y])

# Note the parameters must be passed in the factory and not the callable that is made
dtw_callable = distance_factory(x_numba, y_numba, metric="dtw")
dtw_windowed_callable = distance_factory(x_numba, y_numba, metric="dtw", window=0.2)
edr_callable = distance_factory(x_numba, y_numba, metric="edr")
erp_callable = distance_factory(x_numba, y_numba, metric="erp", epsilon=1.0)
lcss_callable = distance_factory(x_numba, y_numba, metric="lcss")
msm_callable = distance_factory(x_numba, y_numba, metric="msm")
twe_callable = distance_factory(x_numba, y_numba, metric="twe")
wdtw_callable = distance_factory(x_numba, y_numba, metric="wdtw")
wdtw_callable_g2 = distance_factory(x_numba, y_numba, metric="wdtw", g=0.2)
wdtw_callable_g3 = distance_factory(x_numba, y_numba, metric="wdtw", g=0.2)

dtw_numba = dtw_callable(x_numba, y_numba)
dtw_windowed_numba = dtw_windowed_callable(x_numba, y_numba)
edr_numba = edr_callable(x_numba, y_numba)
erp_numba = erp_callable(x_numba, y_numba)
lcss_numba = lcss_callable(x_numba, y_numba)
msm_numba = msm_callable(x_numba, y_numba)
twe_numba = twe_callable(x_numba, y_numba)
wdtw_numba = wdtw_callable(x_numba, y_numba)
wdtw_g2_numba = wdtw_callable_g2(x_numba, y_numba)
wdtw_g3_numba = wdtw_callable_g3(x_numba, y_numba)
```

```python
# Generate the path for each distance
from aeon.distances import distance_alignment_path

dtw_path = distance_alignment_path(x, y, metric="dtw")
dtw_path_windowed = distance_alignment_path(x, y, metric="dtw", window=0.2)
edr_path = distance_alignment_path(x, y, metric="edr")
erp_path = distance_alignment_path(x, y, metric="erp", epsilon=1.0)
lcss_path = distance_alignment_path(x, y, metric="lcss")
msm_path = distance_alignment_path(x, y, metric="msm")
twe_path = distance_alignment_path(x, y, metric="twe")
wdtw_path = distance_alignment_path(x, y, metric="wdtw")
wdtw_path_g2 = distance_alignment_path(x, y, metric="wdtw", g=0.2)
wdtw_path_g3 = distance_alignment_path(x, y, metric="wdtw", g=0.3)
```

```python
# This function does the above but generates a graph from it
from aeon._contrib._plot_path import _plot_alignment, _plot_path

plt_path = _plot_path(x, y, metric="dtw", title="dtw path window 0.2")
plt_path.show()
plt_alignment = _plot_alignment(x, y, metric="dtw", title="dtw alignment window 0.2")
plt_alignment.show()
```
