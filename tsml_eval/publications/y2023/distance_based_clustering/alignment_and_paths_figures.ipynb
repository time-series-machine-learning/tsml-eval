<h1>Paper path and alignment generation</h1>

```python
import warnings

import numpy as np

warnings.filterwarnings("ignore")  # Hide warnings

# dummy time series data
x = np.array(
    [
        0.7553383207,
        0.4460987596,
        1.197682907,
        0.1714334808,
        0.5639929213,
        0.6891222874,
        1.793828873,
        0.06570866314,
        0.2877381702,
        1.633620422,
    ]
)
y = np.array(
    [
        -0.01765193577,
        -1.536784164,
        -0.1413292622,
        -0.7609346135,
        -0.1767363331,
        -2.192007072,
        -0.1933165696,
        -0.4648166839,
        -0.9444888843,
        -0.239523623,
    ]
)
```

```python
# Find the distance between time series
from aeon.distances import distance

dtw = distance(x, y, method="dtw")
dtw_windowed = distance(x, y, method="dtw", window=0.2)
edr = distance(x, y, method="edr")
erp = distance(x, y, method="erp")
lcss = distance(x, y, method="lcss")
msm = distance(x, y, method="msm")
twe = distance(x, y, method="twe")
wdtw = distance(x, y, method="wdtw")
wdtw_g2 = distance(x, y, method="wdtw", g=0.2)
wdtw_g3 = distance(x, y, method="wdtw", g=0.3)
```

```python
# Generate the path for each distance
from aeon.distances import alignment_path

dtw_path = alignment_path(x, y, method="dtw")
dtw_path_windowed = alignment_path(x, y, method="dtw", window=0.2)
edr_path = alignment_path(x, y, method="edr")
erp_path = alignment_path(x, y, method="erp")
lcss_path = alignment_path(x, y, method="lcss")
msm_path = alignment_path(x, y, method="msm")
twe_path = alignment_path(x, y, method="twe")
wdtw_path = alignment_path(x, y, method="wdtw")
wdtw_path_g2 = alignment_path(x, y, method="wdtw", g=0.2)
wdtw_path_g3 = alignment_path(x, y, method="wdtw", g=0.3)
```

```python
# This function does the above but generates a graph from it
from tsml_eval.publications.y2023.distance_based_clustering._plot_path import (
    _plot_alignment,
    _plot_path,
)

plt_path = _plot_path(x, y, method="dtw", title="dtw path window 0.2")
plt_path.show()
plt_alignment = _plot_alignment(x, y, method="dtw", title="dtw alignment window 0.2")
plt_alignment.show()
```
